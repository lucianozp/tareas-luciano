```python
# Detectamos el puntaje gamer de una persona, viendo si es principante, intermedio o pro
puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

suma = puntaje1 + puntaje2

# Evaluamos el resultado según las reglas
if suma < 50:
    print(f"Resultado: Nivel principiante")
elif 50 <= suma < 80:
    print(f"Resultado: Nivel intermedio")
else:
    print(f"Resultado: Nivel pro")
```
