```python
# Números especial del robot de 1 al 100 divisibles por 5 o por 10

for i in range(1, 101):
    if i % 5 == 0:
        # Si es divisible por 5 y por 10 (los que terminan en 0)
        if i % 10 == 0:
            print(f"{i} Bip Bip")
        # Si solo es divisible por 5 (los que terminan en 5)
        else:
            print(f"{i} Bip")
```
